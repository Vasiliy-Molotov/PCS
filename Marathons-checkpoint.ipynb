```python
#загружаем дейтасет

test_raw = pd.read_excel(io = way_import + str('\\Marathons_data.xlsx'))

# test_raw = pd.read_excel(io = way_import + str('\\Marathons_data_2.xlsx'))
# test_raw = test_raw[test_raw['#_did_tasks']==1]

# выберем нужную колонку
# test_raw.columns
# neded_col = 'Чем бы вы хотели профессионально заниматься на горизонте до 2 лет и почему?'
# neded_col ='Что полезного вы ожидаете получить от курса?'
# neded_col ='Расскажите, чем вас заинтересовало участие в курсе?'
# neded_col ='Какими внеучебными / рабочими активностями вы в настоящий момент занимаетесь (спорт / курсы повышения квалификации / изучение языков / участие в кейс-чемпионатах и пр.)?'
neded_col = 'Какие ваши 3 самые любимые дисциплины в ВУЗе?'
cols = [neded_col]
research_df = test_raw[neded_col]
research_df = research_df.reset_index(drop = True)
```

```python
# stopwords для чистки словаря

stopwords = pd.read_csv(way + '\\stop-words_russian_2_ru.txt',header=None)
# stopwords = stopwords.reset_index(drop = True)
stopwords[0][0]='а'

signbase=[',','.',':',';','"','\'','”','“','(',')','«','»','\ufeff','-','\\','/','\xa0', '\xad','+', '\n']

# очистим данные по ней
research_df = pd.DataFrame(test_raw[neded_col].str.lower())
for sign in signbase:
        research_df[neded_col] = research_df[neded_col].str.replace(sign,' ') 
```

```python
# проведем нормализацию словоформ
snowball_stemmer = SnowballStemmer("russian")
for col in cols:
    for i in research_df.index:
        a =  research_df.ix[i,col].split()
    #     for item in a:
    #         if item in list(stopwords[0]):
    #             a.remove(item)
        a[:] = [snowball_stemmer.stem(x) for x in a if x not in list(stopwords[0])]
        research_df.ix[i,col] = ' '.join(a)
        if i % 100 ==0: # контроль прогресса
            print(i)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
word_vectorizer = CountVectorizer(ngram_range=(3,6), analyzer='word')
sparse_matrix = word_vectorizer.fit_transform(research_df[neded_col])
frequencies = sum(sparse_matrix).toarray()[0]
freq = pd.DataFrame(frequencies, index=word_vectorizer.get_feature_names(), columns=['frequency'])
freq = freq.sort(columns = ['frequency'], ascending=False)
```
